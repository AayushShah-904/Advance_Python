```python
import pandas as pd
```

###Read the sales data from all the CSV files

```python
files = ["./Data/sales_april.csv", "./Data/sales_may.csv", "./Data/sales_june.csv"]


df_sales = pd.concat((pd.read_csv(file) for file in files), ignore_index=True)
df_products=pd.read_csv("product_names.csv")
```

```python
df_sales.head()
```

```python
df_products.head()
```

###Calculate the total sales (quantity sold) for each product across all stores and all months.

```python
df_total_avg=df_sales.groupby('Product ID')['Quantity sold'].sum().reset_index()
```

```python
df_total_avg['Avg']=df_total_avg['Quantity sold']/3
df_total_avg['Avg']=df_total_avg['Avg'].round(2)
```

```python
df_total_avg.columns=['Product ID','Total Quantity sold','Average Quantity Sold per Month']
```

```python
df_total_avg.head()
```

```python
df_total_avg.columns
```

###Summary File

```python
summary_df = pd.merge(df_total_avg,df_products,on='Product ID',how='left')
```

```python
summary_df.sort_values('Total Quantity sold',ascending=False)[:5]
```

```python
summary_df.to_csv('sales_summary.csv',index=False)
```
